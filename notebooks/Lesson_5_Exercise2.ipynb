```python
from fastbook import *
```

```python
path = untar_data(URLs.MNIST)
db = DataBlock(
    blocks=(ImageBlock, CategoryBlock), 
    get_items=get_image_files, 
    splitter=RandomSplitter(valid_pct=0.2, seed=42),
    get_y=parent_label)

```

```python
dls = db.dataloaders(path/'training', bs = 16)
learn = cnn_learner(dls, resnet18,pretrained = False,opt_func=SGD, metrics=error_rate)
learn.lr_find(end_lr = 1)
```

```python
learn.fit(10,3e-3)
```

```python
dls = db.dataloaders(path/'training', bs = 64)
learn = cnn_learner(dls, resnet18, pretrained = False,opt_func=SGD, metrics=error_rate)
learn.lr_find(end_lr = 1)
```

```python
learn.fit(10, 2e-2)
```
